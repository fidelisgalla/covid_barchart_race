```python
#importing library
import pandas as pd
import datetime
import threading
from datetime import timedelta
import datetime
import re
```

```python
def daterange(start_date,end_date):
        for n in range(int ((end_date - start_date).days)):
            yield (start_date + timedelta(n)).strftime("%m-%d-%Y")


def urlgenerator(start_date,end_date):
    urls = []
    for date in daterange(start_date,end_date):
        url= 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/{}.csv'.format(date)
        urls.append(url)
    return urls

def getcsv(start_date,end_date):
    urls = urlgenerator(start_date,end_date)
    df_lists=[]
    for i in range(len(urls)):
        df_list = pd.read_csv(urls[i],index_col = False)
        df = df_list[[''.join(re.findall(r'Country/Region|Country_Region',str(df_list.columns))) ,''.join(re.findall(r'Last Update|Last_Update',str(df_list.columns))),'Confirmed','Deaths','Recovered']]
        df_lists.append(df)
    df_final = pd.concat(df_lists,axis=0)
    return df_final
```

```python
class Getdataframe:
    def __init__(self,start,end):
        self.start = start
        self.end = end
        
    def daterange(self):
        for n in range(int ((self.end - self.start).days)):
            yield (self.start + timedelta(n)).strftime("%m-%d-%Y")


    def urlgenerator(self):
        urls = []
        for date in daterange(self.start,self.end):
            url= 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/{}.csv'.format(date)
            urls.append(url)
        return urls

    def getdf(self):
        urls = self.urlgenerator()
        df_lists=[]
        for i in range(len(urls)):
            df_list = pd.read_csv(urls[i],index_col = False)
            df = df_list[[''.join(re.findall(r'Country/Region|Country_Region',str(df_list.columns))) ,''.join(re.findall(r'Last Update|Last_Update',str(df_list.columns))),'Confirmed','Deaths','Recovered']]
            df_lists.append(df)
        df_final = pd.concat(df_lists,axis=0)
        return df_final
    
    def data_cleaner(self,df):
        #df = self.getdf()
        df['Country'] = df['Country/Region'].fillna('') + df['Country_Region'].fillna('')
        df['Date'] = df['Last_Update'].fillna('') + df['Last Update'].fillna('')
        df = df.drop(columns=['Country/Region','Country_Region','Last Update','Last_Update'], axis = 1)
        df['Date'] = pd.to_datetime(df['Date']).dt.date
        df['Date'] = pd.to_datetime(df['Date'])
        df=df[['Date','Country','Confirmed','Deaths','Recovered']]
        return df
    
    def getlatestdata(self):
        try:
            date_today = datetime.datetime.today().date().strftime("%m-%d-%Y")
            url= 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/{}.csv'.format(date_today)
            df_today = pd.read_csv(url,index_col = False)
            df_today = self.data_cleaner(df_today)
        except:
            print('Data yang anda inginkan untuk hari ini tidak ditemukan sehingga data hari sebelumnya akan ditampilkan')
            date_yesterday = (datetime.date.today() - datetime.timedelta(days = 2)).strftime("%m-%d-%Y")
            url =  'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/{}.csv'.format(date_yesterday)
            df_yeserterday = pd.read_csv(url,index_col = False)
            #df_yeserterday = self.data_cleaner(df_yeserterday)
            return df_yeserterday
```
